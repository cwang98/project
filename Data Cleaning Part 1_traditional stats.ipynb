## Data Cleaning

```python
### helper method to find same match
def similarity(x,y):
    x_copy = x.split(" ")
    y_copy = y.split(" ")
    return ((x_copy[0]==y_copy[2])and(x_copy[2]==y_copy[0]))
```

```python
df = pd.read_csv('1617_AllGames_Reg_Advanced.csv', encoding = 'latin1', index_col=0)
df.head(10)
```

```python
### Helper method to find opponent
def find_opponent(teamA,list_new):
    #teamA: row
    MATCH_UP_A = teamA[1]
    GAME_DATE_A = teamA[2]
    for i in range(len(df)):
        if similarity(df.iloc[i][1], MATCH_UP_A) and df.iloc[i][2]==GAME_DATE_A:
            list_new.append(df.iloc[i][:].tolist())
            print("FOUND:", i, "<<", df.iloc[i][0], ">>",df.iloc[i][1])
            break
```

```python
#find opponent for each team in each match and append to a list of list
list_a = []
for a in range(len(df)):
    print("***FINDING:***", a, "<<", df.iloc[a][0], ">>", df.iloc[a][1])
    find_opponent(df.iloc[a],list_a)
```

```python
df_append = pd.DataFrame(list_a, columns = ['OPP_TEAM', 'OPP_MATCH UP', 'OPP_GAME DATE', 'OPP_W/L', 'OPP_MIN', 'OPP_PTS', 'OPP_FGM', 'OPP_FGA','OPP_FG%','OPP_3PM', 'OPP_3PA', 'OPP_3P%','OPP_FTM', 'OPP_FTA', 'OPP_FT%', 'OPP_OREB', 'OPP_DREB', 'OPP_REB', 'OPP_AST', 'OPP_TOV', 'OPP_STL', 'OPP_BLK', 'OPP_PF', 'OPP_+/-'])
```

```python
df_append.head()
```

```python
### Delete columns contain duplicate information 
del df_append['OPP_MATCH UP']
del df_append['OPP_GAME DATE']
del df_append['OPP_MIN']
del df_append['OPP_W/L']
del df_append['OPP_+/-']
```

```python
df_append.head()
```

```python
### Concate opponent information behind own info for each team
new_df = pd.concat([df, df_append], axis=1)
```
