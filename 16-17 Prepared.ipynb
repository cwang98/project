```python
list_teams = ['ATL','BKN','BOS','CHA','CHI','CLE','DAL','DEN','DET','GSW','HOU','IND','LAC','LAL','MEM','MIA',
              'MIL','MIN','NOP','NYK','OKC','ORL','PHI','PHX','POR','SAC','SAS','TOR','UTA','WAS']

df_allPlayers = pd.DataFrame()    
for i in list_teams:
    df_allPlayers = pd.concat([df_allPlayers, df.loc[df['TEAM'] == i]])
```

```python
df_allPlayers.head(30)
```

```python
len(df_allPlayers)
```

```python
for team in list_teams:
    a=0
    for i in range(len(df_allPlayers)):
        if df_allPlayers.iloc[i][1] == team:
            a+=1
            print(team,a)
```
