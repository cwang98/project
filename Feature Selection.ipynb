```python
import pandas as pd
import numpy as numpy
import csv

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('data/1617_AllGames_byTeam.csv', index_col=0)
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
df_ff = pd.read_csv('data/1617_AllGames_byTeam_FF.csv', index_col=0)

del df_ff['TOV']
del df_ff['OPP_TOV']
```

```python
corr = df_ff.corr()

fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            annot=True, cbar=False)
```
